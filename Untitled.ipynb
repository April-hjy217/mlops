```python
import pandas as pd
```

```python
file = "data/raw/yellow_tripdata_2023-01.parquet"
```

```python
df = pd.read_parquet(file)
```

```python
df.head(10)
```

```python
print("Number of columns:", df.shape[1])
```

```python
# Compute duration in minutes
df["duration"] = (
    df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"]
).dt.total_seconds() / 60
```

```python
# Get the standard deviation
std_duration = df["duration"].std()
print("Std of trip durations (min):", round(std_duration, 2))
```

```python
# Filter to [1, 60] minutes
orig_count = len(df)
df_filtered = df[(df.duration >= 1) & (df.duration <= 60)]
filtered_count = len(df_filtered)
```

```python
fraction = filtered_count / orig_count
print("As percentage:", round(fraction * 100, 2), "%")
```

```python
# Cast IDs to strings so DictVectorizer treats them as categories
df["PULocationID"] = df["PULocationID"].astype(str)
df["DOLocationID"] = df["DOLocationID"].astype(str)
```

```python
# Turn into list of dicts
records = df[["PULocationID", "DOLocationID"]].to_dict(orient="records")
```

```python
# Fit the vectorizer
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer(sparse=False)
X = dv.fit_transform(records)
```

```python
# Check dimensionality
print("Feature matrix shape:", X.shape)   
print("Number of one-hot columns:", X.shape[1])
```
